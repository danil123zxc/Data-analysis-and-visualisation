```python
import pandas as pd
```

```python
midwest = pd.read_csv('midwest.csv')
midwest
```

```python
#미성년 인구 백분율 추가
midwest.assign(popchildren = (midwest['poptotal'] - midwest['popadults']) * 100 / midwest['poptotal'])
```

```python
#미성년 인구 백분율이 가장 높은 상위 5개 county
midwest.assign(popchildren = (midwest['poptotal'] - midwest['popadults']) * 100 / midwest['poptotal'])\
.sort_values("popchildren", ascending = False)['county'].head()
```

```python
import numpy as np
```

```python
#분류 변수 추가
midwest.assign(popchildren = lambda x : (x['poptotal'] - x['popadults']) * 100 / x['poptotal'],
              childrencount = lambda x : np.where(x['popchildren'] >= 40, 'large',
                                np.where(x['popchildren'] > 30, 'middle', 'small')))\
.groupby('childrencount')\
.agg(childrencountamount = ('childrencount', 'count'))
```

```python
#아시아인 인구 백분율을 출력
midwest.assign(asianperc = midwest['popasian'] * 100 / midwest['poptotal'])\
.sort_values("asianperc", ascending = False)[['county', 'state', 'asianperc']].head(10)
```
