```python
#변수명 바꾸기
import pandas as pd
```

```python
#데이터 프레임 만들기
df_raw = pd.DataFrame({'var1' : [1, 2, 1], 
                       'var2' : [2, 3, 2]})
df_raw
```

```python
#복사본 만들기
df_new = df_raw.copy()
df_new
```

```python
#변수명 바꾸기
df_new = df_new.rename(columns = {'var2' : 'v2'})
df_new
```
