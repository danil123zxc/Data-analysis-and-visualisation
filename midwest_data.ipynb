```python
import pandas as pd
```

```python
#midwest.csv 파일 불러와 데이터 프레임을 만들겠습니다
data = pd.read_csv("midwest.csv")
data
```

```python
#변수명 수정
data = data.rename(columns = {'poptotal' : 'total', 'popasian': 'asian'})
data
```

```python
#전체 인구 대비 아시아 인구 백분율
data['percasian'] = data['asian'] / data['total'] * 100
data['percasian']
```

```python
#그래프 만들기
data['percasian'].plot.hist()
```

```python
import numpy as np
```

```python
data['asianpop'] = np.where(data['asian'] >= data['asian'].mean(), 'large', 'small')
data['asianpop']
```

```python
#'small'와 'large' 해당하는 지역 빈도 막대 그래프 만들기
data['asianpop'].value_counts().plot.bar(rot=0)
```
